# GIT commands<br>

## Normal workflow<br>
-git status <br>
-git add [filename] <br>
-git commit -m “desc” <br>
-git push <br>

## Repositories<br>
-git diff <br><br>
-git diff -r HEAD <br><br>
	* The -r flag means ‘compare to a particular version’ and HEAD is a shortcut for ‘the most recent commit’<br>
	* You can restrict with git diff -r HEAD path/to/file<br>
-git log <br>
	* to view the log of the project’s history<br>
-git show [first four digits of commits hash]<br>
	* details of a commit<br>
-git add<br>
	* git does not track files by default, use git add at least once to a file so that git starts tracking it. (Git status will let you know which files are not being tracked)<br>
-git clean -f<br>
	* drop files that git is not tracking (in .gitignore) - be very careful <br>

## Undo<br>
-git checkout — . path/to/file<br>
	* will discard the changes that have not yet been staged (haven’t used git add or after reset)<br>
-git reset HEAD path/to/file<br>
	* reset from staging area<br>

Note: think about committing as saving your work, and checking out as loading a saved version. <br>

-git checkout [hash] path/to/file<br>
	* restore an specific version of a file. <br>

## Git Branches<br>
Branching lets you create parallel universes; merging is how you bring them back together. <br>
<br>
-git branch<br>
	*list all the branches in a repository (including master)<br>
-git diff branch-1..branch-2<br>
	*shows the difference between two branches<br>
-git checkout [name of the branch]<br>
	*change branch<br>
-git checkout -b branch-name<br>
	*create a branch (the contents of the new branch are initially identical to the contents of the original.)<br>
-git merge [source] [destination]<br>

## Create new repository<br>
-git init project-name<br>
	*creates a repository from zero<br>
	*do not create a git repository inside another<br>
-git init (in the project’s root directory)<br>
	*existing project into repository<br>
-git clone [url]<br>
	*join a project that is already running<br>

## Remote repository <br>
A typical workflow is that you pull in your collaborators’ work from the remote repository so you have the latest version of everything, do some work yourself, then push your work back to the remote so that your collaborators have access to it. <br>
<br>
-git pull [remote] [branch]<br>
<br>
-git push [remote] [branch]<br>

## Files<br>
.gitignore <br>
	* store a list of wildcard patterns that you don’t want git to keep track of (it could be a name of a folder, or *.csv, etc)<br>
